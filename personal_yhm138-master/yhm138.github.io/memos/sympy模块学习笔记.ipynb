```python
from IPython.display import display, Latex
import sympy as sy
sy.init_printing()

x,y=sy.symbols("x y")
f=x**2+3*x-5
display(Latex(f"$$f(x)={sy.latex(f)}$$"))

xx=3
sy.pprint(f"f({xx})={f.subs({x:xx})}")
yy=4
f1=sy.sqrt(x**2+y**2)

display(Latex(f"$$f_1(x,y)={sy.latex(f1)}$$"))
sy.pprint(f"f1({xx},{yy})={f1.subs({x:xx, y:yy})}")
```

# 一元函数求导

```python
x = sy.Symbol('x')
f=x**2+2*x+1
sy.diff(f,x)
f2 = sy.sin(x)
sy.diff(f2,x)
```

# 多元函数求偏导

```python
y=sy.Symbol('y')
f3=x**2+2*x+y**3
sy.diff(f3,x)
```

```python
sy.diff(f3,y)
```

# 求不定积分

```python
x=sy.Symbol('x')
f=sy.E**x+2*x
sy.integrate(f,x)
```

# 求极限

```python
x=sy.Symbol('x')
f3=(1+1/x)**x
sy.limit(f3,x,sy.oo)
```

```python
f2=(1+x)**(1/x)
sy.limit(f2,x,0)
```

# 展开

```python
x,y,z = sy.symbols("x y z")
f = sy.sin(x+y)
display(Latex(f"$${sy.latex(f)}={sy.latex(sy.expand_trig(f))}$$"))
```
